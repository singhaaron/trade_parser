### import local files | libraries

```python
import numpy as np
import vis
import preprocessor
import process
```

### reformat csv w/proper datatypes & usable columns

```python
relative_path_data = "/sample_data/trade_activity.csv"

trade_csv = preprocessor.frame_data(
    csv_path=relative_path_data, type="TD Ameritrade"
)
```

### process csv

```python
analzyed_csv = process.analyze(trade_csv)
```

```python
%matplotlib inline
```

### Scatter Plot 
#### Dependent Variable: Y: Capital Size
#### Independent Variable X: P&L

```python
vis.scatter_plot(
    np.array(analzyed_csv.loc[:, "p&l"]), np.array(analzyed_csv.loc[:, "cap"])
)
```

### histograms of all data columns include:
######  ticker,profit,loss,capital used,time duration, number of positions(entry & exits)


```python
vis.histogram(df_all=analzyed_csv)
```

#### Linear Regression | Plot Relations
#### Varaibles Used: (P&L | Capital Size | Duration)

```python
vis.linear_regression(
    x=np.array(analzyed_csv.loc[:, "p&l"]), y=np.array(analzyed_csv.loc[:, "cap"])
)
vis.linear_regression(
    x=np.array(analzyed_csv.loc[:, "p&l"]),
    y=np.array(analzyed_csv.loc[:, "time_duration"]),
)
```

```python
vis.plot_line(df_all=analzyed_csv, y="cap", c="black")
vis.plot_line(df_all=analzyed_csv, y="p&l", c="black")
```
